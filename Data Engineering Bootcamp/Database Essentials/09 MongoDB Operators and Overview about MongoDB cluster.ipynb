***Using Operators***

There are three different operators with respect to update – updateOne, updateMany and replaceOne

* If you want to update a particular record first verify using find and then update that data in collection
* Update a single document example-

   <mark>db.orders.updateOne({order_id : 24},{$set : {“$order_status” : ”closed”}})</mark>
        
   <mark>db.orders.find ({ order_id : 24}).pretty()</mark>

* Update more than one document example-

    db.orders.**updateMany**({order_status :”CLOSED”},{$set :{“$order_status” : “closed”}})

***Using Upsert***

* If <u>updateOne()</u>, <u>updateMany()</u>, or <u>replaceOne()</u> includes upsert : true **and** no documents match the specified filter, then the operation creates a new document and inserts it. So let us update orders with order_id that doesn’t exist.
  <mark>db.orders.updateOne({order_id :68884},{$set :{order_id : 68884, “$order_status”  : “UNKNOWN”}},{upsert: true})</mark>

* db.orders.count() will increase by one from the previous count as new document got inserted.
* db.orders.find({order_id : 68884}) will show the new inserted record.

***Using Replace***

      db.orders.replaceOne({“order_id” : 24},{“order_status” : “UNKNOWN”}
      db.orders.find({“order_status” : “UNKNOWN”}).pretty()

### Sharding and Replication in MongoDB
MongoDB is a clustered database.

![image](https://s3.amazonaws.com/info-mongodb-com/_com_assets/media/Architecture%20v4.png)

***Replication***

The replication setting is very much recommended for use in production setting.Replication lets you store copies of data across multiple nodes .This makes data highly available and provides a level of fault tolerance in case a node goes down.

* It can also be used to enhance read operations as clients send read and write requests to different servers.
* A replica set is a group of Mongod instances.In a replica set one node is primary while others are secondary.
* Primary gets all write operations,all data gets replicated from primary to secondary.
* When the primary is not  available, the replica set nominates a secondary as the primary and by default, clients read data from the primary

***Sharding***
* When the volume of data grows huge, sharding in MongoDB provides the data to scale across multiple servers
* Shard are used to store data and each shard is a separate replica set in production environment.
