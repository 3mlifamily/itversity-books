### Launching first EC2 instance
Let us just launch our first EC2 instance and then we will work on all the important concepts associated with it.
* Sign up to https://aws.amazon.com
* You need to share credit card information with them
* Go to AWS Web Console and then to EC2 Dashboard
* Click on Launch Instance
* Go to market place and search for Centos 7
* Configure networking and storage
* Review Security Group
* While launching instance it will ask you to choose existing key pair or new one – first time choose “Create new key pair” option
* Make sure to save private key on your PC with appropriate permissions
* Then go to Dashboard

### EC2 – Elastic Cloud Compute
Now let us understand what is EC2 (Elastic Cloud Compute).
* Memory and CPU
* Instance Store
* Instance is associated with AMI
* With AMI, you will get OS and other additional softwares based on what we have chosen
* EC2 instance is nothing but a (virtual) computer with OS and some softwares running in AWS data center.
* There are different states associated with the computer – running, stop and terminate
* You are charged for compute only when it is running
* When EC2 instance is restarted nothing will be lost, you will go back to where it was unless it fail to restart.
* EC2 instances are ephemeral. When it is stopped contents on instance store will be erased.
* If instance is stopped and started, public ip also will change. To overcome this we can configure elastic ip which will be covered later.
* When terminated every thing on the instance will be gone, EBS volumes will be detached or deleted. To get volumes deleted automatically you need to enable that while creating the instance.

### Networking
Now let us understand the concepts behind networking with respect to AWS.
* We use different ips for different purposes in AWS
* Private ip which does not change until terminated is used to communicate with other EC2 instances with in VPC (Virtual Private Cloud)
* Public ip for external world to communicate with the EC2 instance. Default public ip is free, but it change every time we stop and start  or terminate and create EC2 instance.
* Elastic ip can be used to over come this, but it comes at nominal cost. We can create and then attach or detach elastic ip to any EC2 instance.

### Storage
AWS supports different types of storage – ranging from local to instance to accessible from any where.
* s3 – Simple Cloud Storage Service. This is cheapest of all, accessible from any where with proper authentication and also support offline cheaper options such as Glacier.
* EBS – Elastic Block Storage. Storage which can be attached and detached from EC2 instances. root file system have to be EBS to store operating system files so that EC2 instances can be restarted.
* Instance Store – it is free but ephemeral. It means all the files stored in this type of storage will be lost if the EC2 instance is stopped or terminated. Most of the EC2 instances comes with good amount of free storage of this category.
* Also there are other forms of storage such as EFS (which can be leveraged with AWS EMR)

### Authentication
Let us understand how we can connect to EC2 instance while it is running.
* By default password login is disabled for EC2 instances.
* We need to have Key Pair to connect with out password.
* Follow the instructions if you want to use putty from Windows laptop.
* For Mac or Linux – it is very straight forward. We just need to connect using SSH.
* Key Pairs can be pre-created or can be created while launching the instance.
* One Key Pair can be configured with multiple EC2 instances
* When Key Pair is created, it generates public and private key. Private key will be shared with you only once and you need to preserve it. If you forget, then there is no easy way to connect to instance.
* Corresponding public key will be copied on the instances that are created when we use specific key pair while launching instance.
* Typically the user to which we connect to on EC2 instance will be super user with root permissions.

### Security Groups
While key pairs facilitate you for authenticating to the instance, security group will act as firewall to access applications running on the instance.
* It is logical firewall
* We can apply one security group on multiple EC2 instances to manage rules for specific groups.
* By default, it opens port 22 to enable us for SSH to remote servers
* You need to add rules and control access to the web services running on EC2 instance for external world.
* Even when ports are opened in security group, if there is OS firewall on EC2 instance it might not let you access services on EC2 instance. You might have to add ruless to the OS firewall if it is running.

### Overview of AWS CLI
While we can use AWS Web Console to create instances, you might have to automate the process of creating and managing services on AWS. AWS CLI provides bunch of commands to take care of it.
* It is very easy to setup AWS CLI
* It works on all operating systems
* There are sub commands for each of the services like s3, ec2 etc
* AWS also provide APIs for programming languages like Java, Python etc to perform automation of managing AWS resources.
* You might have to use access key and secret key to access services via AWS CLI

### Overview of AWS Services
While EC2 is PaaS, AWS provide many others “as a service”. They are categorized into different groups.
* DS – Relational Database Service
    * Oracle
    * MySQL and Aurora
    * Red Shift
    * and many more
* Analytics – Whole suite of Big Data and tools related to analytics
    * Elastic Map Reduce – Hadoop and Spark as a service
    * Kinesis – real time data ingestion and streaming
    * Athena – SQL on top of S3
    * and more
* Lambda – for serverless applications
* and more
